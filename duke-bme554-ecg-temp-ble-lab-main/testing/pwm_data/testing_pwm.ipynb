# Pulse Width Modulation Implementation Analysis Report

## Steady-State PWM Output

### Methods

For this part of the PWM lab we were tasked with reading in input from the AIN0 channel of the ADC and map this read DC voltage to scale the maximum brightness of LED2 using a PWM output when a button is pressed (frequency down button in my code). The equation that I used to map the voltage to the period is period  = (pwm.period * (input mV reading/ 3000.0)) which should result in a direct linear relationship of the period and input voltage. Slight variations from this mapping can be due to noise, resolution limits, and/or timing. The function generator was used to test different input voltages and the oscilliscope was used to measure the period duty cycle. I measured various voltages, 0.0 0.5 1.0 1.5 2.0 2.5 and 3.0 V, all at 0.15A, 3 different times to perform robust testing and get more accurate results to mitigate error. The oscilliscope was set to a 25,000 sampling at 2.5k samples per second rate. Once the voltage was mapped to the period, the LED would turn on at the specified duty cycle (brightness) for 5 seconds at the mapped period and if an error occured, it was logged and the device went to the error state.

```python
# ==========================================================
# LED2 PWM Linearity Analysis (Active-Load)
# ==========================================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from scipy.stats import linregress, t

# ----------------------------------------------------------
# 1. Configuration
# ----------------------------------------------------------
# Define your voltage → file mapping (as you already have)
data_groups = {
    0.0:  ['scope_54.csv', 'scope_55.csv', 'scope_56.csv'],
    0.5:  ['scope_36.csv', 'scope_37.csv', 'scope_38.csv'],
    1.0:  ['scope_39.csv', 'scope_40.csv', 'scope_41.csv'],
    1.5:  ['scope_42.csv', 'scope_43.csv', 'scope_44.csv'],
    2.0:  ['scope_45.csv', 'scope_46.csv', 'scope_47.csv'],
    2.5:  ['scope_48.csv', 'scope_49.csv', 'scope_50.csv'],
    3.0:  ['scope_51.csv', 'scope_52.csv', 'scope_53.csv'],
}

# ----------------------------------------------------------
# 2. Helper: Read a single scope CSV
# ----------------------------------------------------------
def read_scope_csv(filepath):
    """Reads Tektronix/Keysight-style scope CSV with two header lines."""
    df = pd.read_csv(filepath, skiprows=2, names=["Time (s)", "Voltage (V)"])
    df["Time (s)"] = pd.to_numeric(df["Time (s)"], errors="coerce")
    df["Voltage (V)"] = pd.to_numeric(df["Voltage (V)"], errors="coerce")
    df = df.dropna()
    return df

# ----------------------------------------------------------
# 3. Helper: Compute duty cycle for active-low PWM
# ----------------------------------------------------------
def compute_duty_cycle_active_low(voltage_array, tol=0.05):
    """
    Compute duty cycle for active-low PWM (LED ON = low voltage).
    Handles nearly constant signals robustly.
    tol: voltage difference to consider signal "flat"
    """
    vmin, vmax = np.min(voltage_array), np.max(voltage_array)
    threshold = (vmin + vmax) / 2

    if (vmax - vmin) < tol:
        # Signal is essentially flat
        if voltage_array[0] >= vmax - tol:
            duty = 0.0   # LED mostly OFF
        else:
            duty = 100.0 # LED mostly ON
    else:
        # Active-low: LED ON when voltage is low → duty = fraction above threshold
        duty = np.mean(voltage_array < threshold) * 100.0

    return np.clip(duty, 0, 100), threshold, vmin, vmax


#  stats on raw voltage !!!!!!
# ---# ----------------------------------------------------------
# 4. Process all data files
# ----------------------------------------------------------
records = []

for vin, file_list in data_groups.items():
    for f in file_list:
        path = Path(f)
        if not path.exists():
            raise FileNotFoundError(f"Missing: {path}")
        df = read_scope_csv(path)
        duty, thresh, vmin, vmax = compute_duty_cycle_active_low(df["Voltage (V)"].values)

        # --- Record everything ---
        records.append({
            "file": f,
            "Input Voltage (V)": vin,
            "Duty (%)": duty,
            "Vmin": vmin,
            "Vmax": vmax,
            "Threshold": thresh
        })

        # --- Optional: detect abnormal duty cycles ---
        if vin == 0.0 and duty > 10:  # expect ~0%, so flag if >10%
            print(f"⚠️  WARNING: Unexpected high duty ({duty:.2f}%) for {f} at {vin} V")

df_all = pd.DataFrame(records)

# Detect outliers compared to the group mean
outlier_threshold = 3  # standard deviations
df_all = df_all.join(
    df_all.groupby("Input Voltage (V)")["Duty (%)"]
    .transform(lambda x: (x - x.mean()).abs() / x.std())
    .rename("zscore")
)

outliers = df_all[df_all["zscore"] > outlier_threshold]

if not outliers.empty:
    print("\n🚨 Potential outlier files:")
    print(outliers[["Input Voltage (V)", "file", "Duty (%)", "zscore"]])


# ----------------------------------------------------------
# 5. Compute averages and standard deviations
# ----------------------------------------------------------
df_summary = (
    df_all.groupby("Input Voltage (V)")
          .agg(mean_duty=("Duty (%)", "mean"),
               std_duty=("Duty (%)", "std"))
          .reset_index()
)

print("Average Duty Cycle per Input Voltage:")
display(df_summary)

# ----------------------------------------------------------
# 6. Linear regression + confidence intervals
# ----------------------------------------------------------
def linear_fit_with_ci(x, y):
    lr = linregress(x, y)
    slope, intercept, r_value, p_value, std_err = lr
    r2 = r_value**2

    n = len(x)
    t_crit = t.ppf(0.975, df=n-2)
    slope_ci = (slope - t_crit * std_err, slope + t_crit * std_err)

    intercept_se = std_err * np.sqrt(1/n + np.mean(x)**2 / np.sum((x - np.mean(x))**2))
    intercept_ci = (intercept - t_crit * intercept_se, intercept + t_crit * intercept_se)

    return slope, intercept, r2, slope_ci, intercept_ci

# Full-range fit (0–3 V)
x_all = df_summary["Input Voltage (V)"]
y_all = df_summary["mean_duty"]
slope_all, intercept_all, r2_all, ci_s_all, ci_i_all = linear_fit_with_ci(x_all, y_all)

# Linear-range fit (0.5–2.5 V)
mask = (x_all >= 0.5) & (x_all <= 2.5)
x_lin, y_lin = x_all[mask], y_all[mask]
slope_lin, intercept_lin, r2_lin, ci_s_lin, ci_i_lin = linear_fit_with_ci(x_lin, y_lin)

# ----------------------------------------------------------
# 7. Plot data and residuals
# ----------------------------------------------------------
plt.figure(figsize=(12,5))

# Left: Linearity plot
plt.subplot(1,2,1)
plt.errorbar(x_all, y_all, yerr=df_summary["std_duty"], fmt='o', capsize=4,
             label="Measured (mean ± std)", color='C0')

xx = np.linspace(0, 3, 100)
yy_all = slope_all * xx + intercept_all
yy_lin = slope_lin * xx + intercept_lin
yy_ideal = xx / 3 * 100

plt.plot(xx, yy_all, 'r-', label=f"Linear fit (full): y={slope_all:.2f}x+{intercept_all:.2f}\nR²={r2_all:.4f}")
plt.plot(xx, yy_lin, '--', color='orange', label=f"Linear fit (0.5–2.5 V): y={slope_lin:.2f}x+{intercept_lin:.2f}\nR²={r2_lin:.4f}")
plt.plot(xx, yy_ideal, ':', color='green', label="Ideal mapping (0 V→0%, 3 V→100%)")

plt.xlabel("Input Voltage (V)")
plt.ylabel("LED2 Brightness Duty Cycle (%)")
plt.title("LED Brightness Linearity (Active-Load PWM)")
plt.legend()
plt.grid(True)

# Right: Residual plot for full-range fit
plt.subplot(1,2,2)
residuals = y_all - (slope_all * x_all + intercept_all)
plt.scatter(x_all, residuals, color='purple', s=80)
plt.axhline(0, color='red', linestyle='--')
for xi, ri in zip(x_all, residuals):
    plt.annotate(f"{ri:.1f}", (xi + 0.02, ri), fontsize=9)
plt.xlabel("Input Voltage (V)")
plt.ylabel("Residuals (%)")
plt.title("Residual Plot (Deviation from Full-Range Fit)")
plt.grid(True)

plt.tight_layout()
plt.show()

# ----------------------------------------------------------
# 8. Print numerical results
# ----------------------------------------------------------
print("Full-range fit (0–3 V):")
print(f"  slope = {slope_all:.4f}, intercept = {intercept_all:.4f}, R² = {r2_all:.5f}")
print(f"  95% CI slope = ({ci_s_all[0]:.4f}, {ci_s_all[1]:.4f})")
print(f"  95% CI intercept = ({ci_i_all[0]:.4f}, {ci_i_all[1]:.4f})\n")

print("Linear-range fit (0.5–2.5 V):")
print(f"  slope = {slope_lin:.4f}, intercept = {intercept_lin:.4f}, R² = {r2_lin:.5f}")
print(f"  95% CI slope = ({ci_s_lin[0]:.4f}, {ci_s_lin[1]:.4f})")
print(f"  95% CI intercept = ({ci_i_lin[0]:.4f}, {ci_i_lin[1]:.4f})")
```

## Analysis

The results from the oscilliscope show that LED2 brightness follows a linear relationship with the input voltage read in from AIN0. The results show that the measured duty cycle closely matches the ideal mapping specified in the assignemnt, with an outlier at 0.0V. Without this data point, the linear relationship can be described with an R² = 1, while with the 0.0V data points the R² = 0.3367, which can be due to the input voltage mapping not going exactly to 0.0 V so the resultant duty cycle is non-zero. This is the reason for the linear fit full line (red) showing a non-linear realtionship across the input voltage range. In my first graph generated, it shows that for the 0V input value, the PWM for the LED behaves as expected, so the error can be due to not doing the analysis on the raw values. 

The residual plot shows the deviations in the output data from the oscilliscope. This graph more clearly shows the error behavior in the 0.0V input mapping which can be due to leakage current and/or PWM driver not reaching a true 0% duty cycle. Overall I would say that the device operates within the desired output range and deviates slightly near the lower limit due to hardware control errors. 

## Sinusoidal Modulation of PWM Output

### Methods

For this part of the PWM lab, we were tasked with modulating the brighness of LED3 to match sampling a 2-second, 10 Hz differential sinusoidal voltage on AIN1 and AIN2 after pressing BUTTON2 (frequency up button in my code). The way I implemented this was to call adc_read_async, which would get 400 sample data points, and after each point was acquired, the callback function would submit a work queue request that would convert the ADC read value to mV, convert that mV value to the PWM period, then set the PWM period for the correct LED with pwm_set_pulse_dt(). To calculate the PWM period I used period = (mV value + 1000.0) / 2000.0 which maps the -1V -> 1V input voltage range to a duty cycle from 0-100%. I used the oscilloscope to concurrently measure my input sinusoidal signal (which is the differential input of AIN1 and AIN2) and the output PWM signal on LED3. The oscilloscope was set to measure the frequency of my low-pass-filtered PWM signal, which I used to be able to calculate the phase difference between the PWM sinusoid and my input signal. I took 4 measurements to make sure I got accurate data (scope 0-3 csv files).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
from scipy.signal import correlate

# Load CSV
data = pd.read_csv('scope_1.csv', skiprows=2)

# Handle duplicate column names
cols = data.columns.tolist()
new_cols = []
counts = {}

for c in cols:
    if c in counts:
        counts[c] += 1
        new_cols.append(f"{c}_{counts[c]}")
    else:
        counts[c] = 1
        new_cols.append(c)

data.columns = new_cols
data = data.apply(pd.to_numeric, errors='coerce')

# Assume first column is time, next two are input and PWM output
time = data.iloc[:, 0].values
input_signal = data.iloc[:, 1].values  # Input sinusoid
pwm_signal = data.iloc[:, 2].values     # PWM output (filtered)

dt = np.mean(np.diff(time))

print("="*60)
print("PWM SIGNAL ANALYSIS")
print("="*60)

# 1. Calculate PWM frequency using FFT
def get_frequency(sig, dt):
    sig_ac = sig - np.mean(sig)
    n = len(sig_ac)
    fft_vals = fft(sig_ac)
    freqs = fftfreq(n, dt)
    
    pos_mask = freqs > 0
    freqs_pos = freqs[pos_mask]
    fft_mag = np.abs(fft_vals[pos_mask])
    
    peak_idx = np.argmax(fft_mag)
    return freqs_pos[peak_idx]

input_freq = get_frequency(input_signal, dt)
pwm_freq = get_frequency(pwm_signal, dt)

print(f"\nInput signal frequency:  {input_freq:.3f} Hz")
print(f"PWM output frequency:    {pwm_freq:.3f} Hz")
print(f"Expected frequency:      10.000 Hz")
print(f"PWM frequency error:     {abs(pwm_freq - 10.0):.3f} Hz ({abs(pwm_freq - 10.0)/10.0*100:.2f}%)")

# 2. Calculate phase lag using cross-correlation
s1 = input_signal - np.mean(input_signal)
s2 = pwm_signal - np.mean(pwm_signal)

corr = correlate(s2, s1, mode='full')
lags = np.arange(-len(s1) + 1, len(s1))
max_corr_idx = np.argmax(corr)
lag_samples = lags[max_corr_idx]
time_lag = lag_samples * dt

phase_lag_deg = (time_lag * input_freq * 360) % 360
if phase_lag_deg > 180:
    phase_lag_deg -= 360

print(f"\nPhase lag:               {phase_lag_deg:.2f}°")
print(f"Time lag:                {time_lag*1e6:.2f} μs")
print(f"Signal period:           {1/input_freq*1e3:.2f} ms")

# Plot
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))

# Full signal view
colors = ['gold', 'limegreen']
labels = ['Input Signal (AIN1-AIN2)', 'PWM Output (LED3, filtered)']
ax1.plot(time * 1e3, input_signal, color=colors[0], label=labels[0], linewidth=1.5)
ax1.plot(time * 1e3, pwm_signal, color=colors[1], label=labels[1], linewidth=1.5)
ax1.set_xlabel('Time (ms)')
ax1.set_ylabel('Voltage (V)')
ax1.set_title('Input Sinusoid and PWM Output')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Zoomed view to see phase lag (show ~2 cycles)
zoom_time = 2 / input_freq  # 2 periods
zoom_mask = time < zoom_time
ax2.plot(time[zoom_mask] * 1e3, input_signal[zoom_mask], 
         color=colors[0], label=labels[0], linewidth=2)
ax2.plot(time[zoom_mask] * 1e3, pwm_signal[zoom_mask], 
         color=colors[1], label=labels[1], linewidth=2)
ax2.set_xlabel('Time (ms)')
ax2.set_ylabel('Voltage (V)')
ax2.set_title(f'Zoomed View - Phase Lag: {phase_lag_deg:.2f}°')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# Load CSV
data = pd.read_csv('scope_0.csv', skiprows=2)

# Handle duplicate column names
cols = data.columns.tolist()
new_cols = []
counts = {}

for c in cols:
    if c in counts:
        counts[c] += 1
        new_cols.append(f"{c}_{counts[c]}")
    else:
        counts[c] = 1
        new_cols.append(c)

data.columns = new_cols
data = data.apply(pd.to_numeric, errors='coerce')

# Extract signals
time = data.iloc[:, 0].values
input_signal = data.iloc[:, 1].values
pwm_output = data.iloc[:, 2].values

dt = np.mean(np.diff(time))

print("="*70)
print("PWM OUTPUT WAVE CHARACTERISTICS")
print("="*70)

# Detect PWM pulses (find rising edges)
# The PWM signal should have rapid transitions between high and low
pwm_diff = np.diff(pwm_output)
threshold = np.std(pwm_diff) * 2  # Adaptive threshold

# Find rising edges (pulse starts)
rising_edges = np.where(pwm_diff > threshold)[0]

# Find falling edges (pulse ends)
falling_edges = np.where(pwm_diff < -threshold)[0]

if len(rising_edges) > 1:
    # Calculate PWM period (time between pulses)
    pwm_periods = np.diff(time[rising_edges])
    avg_pwm_period = np.mean(pwm_periods)
    pwm_frequency = 1 / avg_pwm_period
    
    print(f"\nPWM Switching Characteristics:")
    print(f"  Number of PWM pulses detected: {len(rising_edges)}")
    print(f"  Average PWM period: {avg_pwm_period*1e6:.2f} μs")
    print(f"  PWM switching frequency: {pwm_frequency:.2f} Hz")
    print(f"  PWM period std dev: {np.std(pwm_periods)*1e6:.2f} μs")
    
    # Calculate duty cycle over time
    # For each pulse, find its duty cycle
    duty_cycles = []
    pulse_times = []
    
    for i, rise_idx in enumerate(rising_edges[:-1]):
        # Find the next falling edge after this rising edge
        next_falls = falling_edges[falling_edges > rise_idx]
        if len(next_falls) == 0:
            continue
        fall_idx = next_falls[0]
        
        # Find next rising edge to get full period
        if i + 1 < len(rising_edges):
            next_rise_idx = rising_edges[i + 1]
            period = time[next_rise_idx] - time[rise_idx]
            on_time = time[fall_idx] - time[rise_idx]
            
            if period > 0:
                duty_cycle = (on_time / period) * 100
                duty_cycles.append(duty_cycle)
                pulse_times.append(time[rise_idx])
    
    if len(duty_cycles) > 0:
        duty_cycles = np.array(duty_cycles)
        pulse_times = np.array(pulse_times)
        
        print(f"\nDuty Cycle Modulation:")
        print(f"  Min duty cycle: {np.min(duty_cycles):.2f}%")
        print(f"  Max duty cycle: {np.max(duty_cycles):.2f}%")
        print(f"  Mean duty cycle: {np.mean(duty_cycles):.2f}%")
        print(f"  Duty cycle range: {np.max(duty_cycles) - np.min(duty_cycles):.2f}%")
        
        # Calculate how duty cycle changes over one input period
        input_period = 0.1  # 10 Hz = 0.1s period
        
        # Find duty cycles in different quarters of the input sine wave
        quarters = []
        for q in range(4):
            start_time = q * input_period / 4
            end_time = (q + 1) * input_period / 4
            mask = (pulse_times >= start_time) & (pulse_times < end_time)
            if np.any(mask):
                avg_dc = np.mean(duty_cycles[mask])
                quarters.append(avg_dc)
        
        if len(quarters) == 4:
            print(f"\nDuty Cycle by Input Phase (first cycle):")
            print(f"  0° - 90°:   {quarters[0]:.2f}%")
            print(f"  90° - 180°: {quarters[1]:.2f}%")
            print(f"  180° - 270°: {quarters[2]:.2f}%")
            print(f"  270° - 360°: {quarters[3]:.2f}%")

# Visualization
fig, axes = plt.subplots(2, 1, figsize=(14, 8))

# Plot 1: Full signal with PWM pulses highlighted
ax1 = axes[0]
ax1.plot(time * 1e3, input_signal, color='gold', label='Input Signal', linewidth=2, alpha=0.8)
ax1.plot(time * 1e3, pwm_output, color='limegreen', label='PWM Output', linewidth=1)

# Highlight some rising edges
if len(rising_edges) > 0:
    sample_edges = rising_edges[::max(1, len(rising_edges)//20)]  # Show ~20 markers
    ax1.scatter(time[sample_edges] * 1e3, pwm_output[sample_edges], 
                color='red', s=30, zorder=5, label='PWM Pulse Starts', alpha=0.6)

ax1.set_xlabel('Time (ms)')
ax1.set_ylabel('Voltage (V)')
ax1.set_title('Input Sinusoid and PWM Output')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Plot 2: Zoomed view showing PWM switching
zoom_duration = 0.2  # 200ms window
zoom_start = 0.5  # Start at 500ms
zoom_mask = (time >= zoom_start) & (time < zoom_start + zoom_duration)

ax2 = axes[1]
ax2.plot(time[zoom_mask] * 1e3, input_signal[zoom_mask], 
         color='gold', label='Input Signal', linewidth=2)
ax2.plot(time[zoom_mask] * 1e3, pwm_output[zoom_mask], 
         color='limegreen', label='PWM Output', linewidth=1.5)

# Mark pulses in this window
zoom_edges = rising_edges[(time[rising_edges] >= zoom_start) & 
                          (time[rising_edges] < zoom_start + zoom_duration)]
if len(zoom_edges) > 0:
    ax2.scatter(time[zoom_edges] * 1e3, pwm_output[zoom_edges], 
                color='red', s=50, zorder=5, label='PWM Pulses')

ax2.set_xlabel('Time (ms)')
ax2.set_ylabel('Voltage (V)')
ax2.set_title(f'Zoomed View - PWM Switching Detail ({zoom_duration*1e3:.0f} ms window)')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## Analysis

My results show that there is a phase lag of about -170 degrees and a time lag of 52000.00 μs. This system delay can be due to my 1:1 mapping of the input phase read in with adc_read_async() and setting the PWM with pwm_set_pulse_dt(). Another reason for the lag could be from the ADC conversion time, callback execution, and the PWM update rate, as mentioned earlier. Another way I could have implemented this would be by taking a few data points (3-5) and averaging that data and then setting the PWM with the average. This would also mitigate the need to have to cap the input adc_read_async() values that were read in out of range (>1000 mV & < -1000mV). A couple other ways to mitigate the latency would be to increase the ADC aquition rate or write faster code for the callback function. 

The duty cycle outputted and measured shows the brightness of the LED, with a higher input voltage resulting in a brighter LED and a higher duty cycle. My results look pretty accurate in terms of mapping the input sinusodal frequency to to corresponding output PWM, while also changing in real time based on the input read in. All of my samples read in show the correct updated PWM with a range measure from 11.11% to 91.67%. Since this is not from 0-100% duty cycle the error is probably due to the 1:1 mapping and not being able to change the PWM fast enough to be able to read on the oscilliscope, or from the number of oscilliscope data points being measured.
